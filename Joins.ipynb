```
#Dataframes 
person = spark.createDataFrame([ (0, "Bill Chambers", 0, [100]),   
                                 (1, "Matei Zaharia", 1, [500, 250, 100]),     
                                 (2, "Michael Armbrust", 1, [250, 100])])\
             .toDF("id", "name", "graduate_program", "spark_status") 
graduateProgram = spark.createDataFrame([ (0, "Masters", "School of Information", "UC Berkeley"),
                                          (2, "Masters", "EECS", "UC Berkeley"), 
                                           (1, "Ph.D.", "EECS", "UC Berkeley")])\
                   .toDF("id", "degree", "department", "school") 
sparkStatus = spark.createDataFrame([  (500, "Vice President"), 
                                       (250, "PMC Member"), 
                                       (100, "Contributor")])\
                     .toDF("id", "status")

  
```

```
#Inner Joins : Note : Inner Join is Default joining methods in spark dataframe 
joinexpression = person["graduate_program"] == graduateProgram["id"]
person.join(graduateProgram,joinexpression).show()
```

```
# How to explicit mention the joining type 

joinexpression = person["graduate_program"] ==graduateProgram["id"]
joiningType  = "inner"
person.join(graduateProgram,joinexpression,joiningType).show()
```

```
#Outer Joining - Which provides both Matched and Unmatched records.
joinexpression = person["graduate_program"] == graduateProgram["id"]
joiningType_out  = "outer"
person.join(graduateProgram,joinexpression,joiningType_out).show()
```

```
#Left Left Outer Joining - Which provides both Matched and Unmatched records.
joinexpression = person["graduate_program"] == graduateProgram["id"]
joiningType_out  = "left_outer"
person.join(graduateProgram,joinexpression,joiningType_out).show()
```

```
#Outer Right Joining - Which provides both Matched and Unmatched records.
joinexpression = person["graduate_program"] == graduateProgram["id"]
joiningType_out  = "right_outer"
person.join(graduateProgram,joinexpression,joiningType_out).show()
```

```
#Outer Right Joining - Which provides both Matched and Unmatched records.
joinexpression = person["graduate_program"] == graduateProgram["id"]
joiningType_out  = "left_semi"
person.join(graduateProgram,joinexpression,joiningType_out).show()
```

```
#Outer Right Joining - Which provides both Matched and Unmatched records.
2
joinexpression = person["graduate_program"] == graduateProgram["id"]
3
joiningType_out  = "right_semi"
4
person.join(graduateProgram,joinexpression,joiningType_out).show()
```

```
#Outer Right Joining - Which provides both Matched and Unmatched records.
joinexpression = person["graduate_program"] == graduateProgram["id"]
joiningType_out  = "right_outer"
person.join(graduateProgram,joinexpression,joiningType_out).show()
```

```
#left Anti Join 

joinexpression = person["graduate_program"] == graduateProgram["id"]
joiningType_out  = "left_anti"
person.join(graduateProgram,joinexpression,joiningType_out).show()
```

```
joinexpression = person["graduate_program"] == graduateProgram["id"]
joiningType_out  = "right_anti"
person.join(graduateProgram,joinexpression,joiningType_out).show()
```
