```
from pyspark import SQLContext
from pyspark import HiveContext 
a = [("karthikeyan",37)]
rdd = sc.parallelize(a)
schema = StructType([StructField("name",StringType()),
                     StructField("age",StringType())
                    ]
                   )
df = spark.createDataFrame(rdd,schema)
df.show()
```
