```
from pyspark import SQLContext
from pyspark import HiveContext
from pyspark.sql.functions import struct
rawdata = [[1,1000,'Apple',0.76],
           [2,1000,'Apple',0.11],
           [1,2000,'oragange',0.98],
           [1,3000,'Banana',0.24],
           [2,2000,'Banana',0.99]
          ]
df =spark.createDataFrame(rawdata).toDF('key','itemkey','itemname','score')
df.show()
df.groupBy('key')\
  .agg(sort_array(collect_list(struct('score','itemkey','itemname'))))\
  .toDF('key','collecttion').show()
```
